```python
import pandas as pd
path = r"C:\Users\tanch\Documents\NTU\NTU Year 4\Semester 1\CZ4125 - Developing Data Products\Assignments\Team Assignment - book recommendation\data\consumer-reviews-of-amazon-products.csv"
data = pd.read_csv(path, usecols = ["id", "reviews.title", "reviews.text"])
data
```

```python
def get_mock_review_json(row):
    output = {}
    output["ISBN"] = row.id
    output["summary"]  = row["reviews.title"]
    output["genre"]  = str(row["reviews.title"]).split()
    output["reviews"]  = [row["reviews.text"]]
    return output
```

```python
import json
path = r"C:\Users\tanch\Documents\NTU\NTU Year 4\Semester 1\CZ4125 - Developing Data Products\Assignments\Team Assignment - book recommendation\data\test.json"
for i in range(len(data)):
    output = get_mock_review_json(data.iloc[i])
    with open(path, mode='a') as f:
        json.dump(output, f)
        f.write('\n')
```
