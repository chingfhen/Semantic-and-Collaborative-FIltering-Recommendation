## find documents

```python
# find by ids
recommended_ids = [ObjectId("63270bdd09cd13fff5fcb5e3"),ObjectId("63270bdd09cd13fff5fcb5c7"),ObjectId("63270bdd09cd13fff5fcb5de")]
for d in collection.find({"_id":{"$in":recommended_ids}}):
    print(d)
```

```python
# find any that contain "kindle" or "perfect" keyword
for d in collection.find({"$or":[{"keywords":"kindle"},{"keywords":"perfect"}]}):
    print(d)
```

```python
# find any that contain "kindle" or "perfect" keyword  AND genre "Worth"
for d in collection.find({"$and":
                          [{"genre":"Worth"}, {"$or":[{"keywords":"kindle"},{"keywords":"perfect"}]}]}):
    print(d)
```
